```python
def get_category(age):
    if age == None or np.isnan(age): return 'NA'
    if age <= 5: return 'Baby'
    if age <= 12: return 'Child'
    if age <= 18: return 'Teenager'
    if age <= 25: return 'Student'
    if age <= 35: return 'Young Adult'
    if age <= 60: return 'Adult'
    return 'Elderly'
```

```python
titanic_pdf['Age_Category'] = titanic_pdf['Age'].map(get_category)
titanic_pdf.head(10)
```

```python
from pyspark.sql.functions import udf, col
from pyspark.sql.types import StringType

udf_get_category = udf(get_category, StringType())
```

```python
titanic_sdf_filled_01 = titanic_sdf.withColumn('Age_Category', udf_get_category(col('Age')))
titanic_sdf_filled_01.show()
```

```python
from pyspark.sql.functions import when

titanic_sdf_filled_02 = titanic_sdf_filled.withColumn('Age_category',
    when(col('age') <= 5, 'Baby')
    .when(col('age') <= 12, 'Child')
    .when(col('age') <= 18, 'Teenage')
    .when(col('age') <= 25, 'Student')
    .when(col('age') <= 35, 'Young Adult')
    .when(col('age') <= 60, 'Adult')
    .when(col('age').isNull(), 'NA')
    .otherwise('Elderly')
)
titanic_sdf_filled_02.show()

'''
SELECT t.*, 
    CASE WHEN age <= 5 THEN 'Baby'
    WHEN age <= 12 THEN 'Child'
    WHEN age <= 18 THEN 'Teenage'
    WHEN age <= 25 THEN 'Student'
    WHEN age <= 35 THEN 'Young Adult'
    WHEN age <= 60 THEN 'Adult'
    WHEN age is null THEN 'NA'
    ELSE 'Elderly' END
from titanic_sdf_filled t
'''
```
